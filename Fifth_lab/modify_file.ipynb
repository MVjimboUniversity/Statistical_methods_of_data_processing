# Обработка временного ряда

```python
import pandas as pd
```

```python
data = pd.read_csv("weather.csv", sep=";", index_col=False)
data.head()
```

```python
data.info()
```

```python
data["Date"] = pd.to_datetime(data["Местное время в Санкт-Петербурге"].copy(deep=True)).dt.date
```

```python
data = data.drop("Местное время в Санкт-Петербурге", axis=1)
```

```python
data_temperature = data[["Date", "T"]].copy(deep=True)
```

```python
data_temperature.head()
```

```python
data_temperature_grouped = data_temperature.groupby(by="Date", as_index=False).mean()
```

```python
data_temperature_grouped.to_csv("weather_grouped.csv", index=False)
```
